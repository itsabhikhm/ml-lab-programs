#### <p style="text-align: center;">Implementation of Stochastic Gradient Descent Based Delta Learning Rule</p>

Implement the Stochastic gradient descent rule weight updation method, also use sigmoid activation function, consider appropiate numbers of input neuron and output layer neuron only. Use the given Dataset for training and testing.

##### Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error 
```

##### Defining Inputs and Outputs

```python
df = pd.read_excel('datasets/Lab 8 ML Dataset - Stochastic Gradient Descent(Train).xlsx')
df
```

```python
def sigmoid(net):
    return 1/(1+np.exp(-net))
```

```python
# initialize random weights
W0=0.1
W1=0.2
W2=0.3
W3=0.4
W4=0.5

eta=0.1

for i in range(4):
    for index, current_row in df.iterrows():
        
        net= W0*1+W1*current_row['X1'] + W2*current_row['X2'] + W3*current_row['X3'] + W4*current_row['X4']
        o = sigmoid(net)     
        f_dash_net = o*(1-o) # f'(net) for sigmoid function
        del_w = f_dash_net*(current_row['Y']-o)  # f'(net)*(t-o)

        W0=W0+eta*del_w*1          
        W1=W1+eta*del_w*current_row['X1']              
        W2=W1+eta*del_w*current_row['X2']
        W3=W3+eta*del_w*current_row['X3']               
        W4=W4+eta*del_w*current_row['X4']
print(f"W0 = {W0}")
print(f"W1 = {W1}")
print(f"W2 = {W2}")
print(f"W3 = {W3}")
print(f"W4 = {W4}")
```

Predicting on test Dataset

```python
df2 = pd.read_excel('datasets/Lab 8 ML Dataset - Stochastic Gradient Descent(Test).xlsx')
df2
```

```python
result = []
y_pred = []

for index, row in df2.iterrows():
    net=W0*1+W1*row['X1'] + W2*row['X2'] + W3*row['X3'] + W4*row['X4']
    output = sigmoid(net)
    result.append(output)
    if output > 0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
print(result)
print(y_pred)
```
