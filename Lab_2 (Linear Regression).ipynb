```python
import matplotlib.pyplot as plt
import numpy as np
```

##### Function for finding B0 and B1

```python
def Coefficients (x,y):
  mean_x = np.mean(x)
  mean_y = np.mean(y)
  numerator,dinominator = 0,0
  for i in range(len(x)):
    numerator = numerator + ((x[i] - mean_x)*(y[i] - mean_y))
    dinominator = dinominator + (x[i]-mean_x)**2
  B1 = numerator/dinominator
  B0 = mean_y - B1*mean_x
  return B0,B1
```

##### Function for Predicting Values

```python
def predict(B0,B1,x):
  y_pred = []
  for i in range(len(x)):
    y_pred.append(B0 + B1*x[i])
  return y_pred
```

##### Function to plot linear regression line

```python
def plot(x,y,y_pred):
  plt.figure(figsize=(4, 3))
  plt.scatter(x,y)
  plt.plot(x,y_pred)
  plt.show()
```

#### RMSE

```python
def RMSE(y,y_pred):
  T1 = 0 # Numerator = 0
  for i in range(len(y)):
    T1 = T1 + (y_pred[i] - y[i])**2
  MSE =  T1 / len(y)
  RMSE = MSE**0.5
  return RMSE
```

```python
# Dataset 1
x = [1,2,4,3,5]
y = [1,3,3,2,5]
B0,B1 = Coefficients(x,y)
y_pred = predict(B0,B1,x)
print("Dataset 1 : ---------------------------------")
print("B0 and B1 :",B0,B1)
print("Actual y :",y)
print("Predicted y :",y_pred)
print("RMSE :",RMSE(y,y_pred))
plot(x,y,y_pred)
```

```python
# Dataset 2
Height_x = [100,110,120,130,140,150,160,170,180,190,200,210,220,230,240,250]
Weight_y = [50.1,55.1,60.1,65.1,70.1,75.1,80.1,85.1,90.1,95.1,100.1,105.1,110.1,115.1,120.1,125.1]
B0,B1 = Coefficients(Height_x,Weight_y)
y_pred = predict(B0,B1,Height_x)
print("Dataset 2 : ---------------------------------")
print("B0 and B1 :",B0,B1)
print("Actual y :",Weight_y)
print("Predicted y :",y_pred)
print("RMSE :",RMSE(Weight_y,y_pred))
plot(Height_x,Weight_y,y_pred)
```
